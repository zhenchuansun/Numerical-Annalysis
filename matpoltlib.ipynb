## matpoltlib 组成

1. 脚本层   负责生成图形与坐标系
2. 美工曾   提供了绘制图形的元素时的给各种功能，例如，绘制标题、轴标签、坐标刻度等
3. 后端层   定义了三个基本类， FigureCanvas（图层画布类），它提供了绘图所需的画布，其次是 Renderer（绘图操作类），它提供了在画布上进行绘图的各种方法，最后是 Event（事件处理类），它提供了用来处理鼠标和键盘事件的方法

## Matplotlib.pyplot接口
| 函数名称       | 描述                           |
| -------------- | ------------------------------ |
| `plot`         | 绘制折线图或散点图             |
| `scatter`      | 绘制散点图                     |
| `bar`          | 绘制条形图                     |
| `barh`         | 绘制水平条形图                 |
| `hist`         | 绘制直方图                     |
| `boxplot`      | 绘制箱线图                     |
| `pie`          | 绘制饼图                       |
| `imshow`       | 显示图像                       |
| `contour`      | 绘制等高线图                   |
| `contourf`     | 绘制填充等高线图               |
| `pcolormesh`   | 绘制伪彩色图                   |
| `violinplot`   | 绘制小提琴图                   |
| `errorbar`     | 绘制误差条形图                 |
| `stackplot`    | 绘制堆叠面积图                 |
| `quiver`       | 绘制矢量图                     |
| `streamplot`   | 绘制流线图                     |
| `step`         | 绘制阶梯图                     |

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 折线图
plt.figure()
plt.plot(x, y)
plt.title('Line chart ')# 折线图
plt.show()

# 散点图
plt.figure()
plt.scatter(x, y)
plt.title('Scatter plot')
plt.show()

# 条形图
categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.figure()
plt.bar(categories, values)
plt.title('bar chart')
plt.show()

# 直方图
data = np.random.randn(1000)
plt.figure()
plt.hist(data, bins=30)
plt.title('histogram')
plt.show()

# 箱线图
data = [np.random.randn(100) for _ in range(4)]
plt.figure()
plt.boxplot(data)
plt.title('Boxplot')
plt.show()

# 饼图
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie charts')
plt.show()
```

## Image函数

| 函数名称       | 功能                           |
| -------------- | ------------------------------ |
| `imshow`       | 显示二维数组作为图像           |
| `imsave`       | 将数组保存为图像文件           |
| `imread`       | 从文件读取图像并返回数组       |
| `matshow`      | 显示矩阵数据的图像表示         |
| `figimage`     | 在图形中添加图像               |

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg

# 示例 1：使用 imshow 显示随机生成的二维数组作为图像
def example_imshow():
    data = np.random.rand(10, 10)
    plt.imshow(data, cmap='viridis')
    plt.colorbar()  # 添加颜色条
    plt.title('A randomly generated two-dimensional array image')
    plt.show()

# 示例 2：使用 imsave 保存图像
def example_imsave():
    data = np.random.rand(10, 10)
    plt.imsave('random_image.png', data, cmap='viridis')
    print("图像已保存为 random_image.png")

# 示例 3：使用 imread 读取图像文件并显示
def example_imread():
    img = mpimg.imread('random_image.png')
    plt.imshow(img)
    plt.title('show image')
    plt.axis('off')  # 关闭坐标轴
    plt.show()

# 示例 4：使用 matshow 显示矩阵数据的图像表示
def example_matshow():
    data = np.random.rand(10, 10)
    plt.matshow(data, cmap='viridis')
    plt.colorbar()  # 添加颜色条
    plt.title('Graphical representation of matrix data')
    plt.show()

# 示例 5：使用 figimage 在图形中添加图像
def example_figimage():
    data = np.random.rand(10, 10)
    fig = plt.figure()
    fig.figimage(data, cmap='viridis')
    plt.title('在图形中添加图像')
    plt.show()

def main():
    print("选择一个示例运行:")
    print("1. 使用 imshow 显示随机生成的二维数组作为图像")
    print("2. 使用 imsave 保存图像")
    print("3. 使用 imread 读取图像文件并显示")
    print("4. 使用 matshow 显示矩阵数据的图像表示")
    print("5. 使用 figimage 在图形中添加图像")
    choice = input("请输入选择的编号 (1-5): ")

    if choice == '1':
        example_imshow()
    elif choice == '2':
        example_imsave()
    elif choice == '3':
        example_imread()
    elif choice == '4':
        example_matshow()
    elif choice == '5':
        example_figimage()
    else:
        print("无效的选择，请输入 1 到 5 之间的数字。")

if __name__ == "__main__":
    main()
```

## Axis函数

| 函数名称       | 功能                           |
| -------------- | ------------------------------ |
| `axis`         | 获取或设置坐标轴的属性         |
| `xlabel`       | 设置 x 轴标签                  |
| `ylabel`       | 设置 y 轴标签                  |
| `xlim`         | 获取或设置 x 轴的范围          |
| `ylim`         | 获取或设置 y 轴的范围          |
| `xticks`       | 获取或设置 x 轴的刻度          |
| `yticks`       | 获取或设置 y 轴的刻度          |
| `grid`         | 显示或隐藏网格线               |
| `legend`       | 显示图例                       |
| `title`        | 设置图表标题                   |

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例 1：使用 axis 获取或设置坐标轴的属性
def example_axis():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.axis([0, 10, -1, 1])  # 设置坐标轴范围
    plt.title('使用 axis 设置坐标轴范围')
    plt.show()

# 示例 2：使用 xlabel 和 ylabel 设置轴标签
def example_labels():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.xlabel('时间 (s)')
    plt.ylabel('振幅')
    plt.title('使用 xlabel 和 ylabel 设置轴标签')
    plt.show()

# 示例 3：使用 xlim 和 ylim 获取或设置轴的范围
def example_limits():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.xlim(0, 10)  # 设置 x 轴范围
    plt.ylim(-1, 1)  # 设置 y 轴范围
    plt.title('使用 xlim 和 ylim 设置轴的范围')
    plt.show()

# 示例 4：使用 xticks 和 yticks 获取或设置轴的刻度
def example_ticks():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.xticks(np.arange(0, 11, 1))  # 设置 x 轴刻度
    plt.yticks(np.arange(-1, 1.5, 0.5))  # 设置 y 轴刻度
    plt.title('使用 xticks 和 yticks 设置轴的刻度')
    plt.show()

# 示例 5：使用 grid 显示或隐藏网格线
def example_grid():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.grid(True)  # 显示网格线
    plt.title('使用 grid 显示网格线')
    plt.show()

# 示例 6：使用 legend 显示图例
def example_legend():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    plt.plot(x, y1, label='sin(x)')
    plt.plot(x, y2, label='cos(x)')
    plt.legend()  # 显示图例
    plt.title('使用 legend 显示图例')
    plt.show()

# 示例 7：使用 title 设置图表标题
def example_title():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.title('使用 title 设置图表标题')
    plt.show()

def main():
    print("选择一个示例运行:")
    print("1. 使用 axis 获取或设置坐标轴的属性")
    print("2. 使用 xlabel 和 ylabel 设置轴标签")
    print("3. 使用 xlim 和 ylim 获取或设置轴的范围")
    print("4. 使用 xticks 和 yticks 获取或设置轴的刻度")
    print("5. 使用 grid 显示或隐藏网格线")
    print("6. 使用 legend 显示图例")
    print("7. 使用 title 设置图表标题")
    choice = input("请输入选择的编号 (1-7): ")

    if choice == '1':
        example_axis()
    elif choice == '2':
        example_labels()
    elif choice == '3':
        example_limits()
    elif choice == '4':
        example_ticks()
    elif choice == '5':
        example_grid()
    elif choice == '6':
        example_legend()
    elif choice == '7':
        example_title()
    else:
        print("无效的选择，请输入 1 到 7 之间的数字。")

if __name__ == "__main__":
    main()
```

## figure函数

| 函数名称       | 功能                           |
| -------------- | ------------------------------ |
| `figure`       | 创建一个新的图形               |
| `add_subplot`  | 向图形中添加子图               |
| `savefig`      | 将图形保存到文件               |
| `clf`          | 清除当前图形                   |
| `close`        | 关闭图形窗口                   |
| `gcf`          | 获取当前图形                   |
| `gca`          | 获取当前子图                   |
| `suptitle`     | 设置图形的总标题               |
| `subplots`     | 创建一个包含子图的图形         |
| `tight_layout` | 自动调整子图参数以填充图形区域 |

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例 1：使用 figure 创建一个新的图形
def example_figure():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure()
    plt.plot(x, y)
    plt.title('使用 figure 创建一个新的图形')
    plt.show()

# 示例 2：使用 add_subplot 向图形中添加子图
def example_add_subplot():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    fig = plt.figure()
    ax1 = fig.add_subplot(2, 1, 1)
    ax1.plot(x, y1)
    ax1.set_title('子图 1: sin(x)')
    ax2 = fig.add_subplot(2, 1, 2)
    ax2.plot(x, y2)
    ax2.set_title('子图 2: cos(x)')
    plt.show()

# 示例 3：使用 savefig 将图形保存到文件
def example_savefig():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure()
    plt.plot(x, y)
    plt.title('使用 savefig 将图形保存到文件')
    plt.savefig('saved_figure.png')
    print("图形已保存为 saved_figure.png")

# 示例 4：使用 clf 清除当前图形
def example_clf():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure()
    plt.plot(x, y)
    plt.title('图形将在 2 秒后清除')
    plt.show()
    plt.clf()
    print("图形已清除")

# 示例 5：使用 close 关闭图形窗口
def example_close():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure()
    plt.plot(x, y)
    plt.title('图形将在 2 秒后关闭')
    plt.show()
    plt.close()
    print("图形已关闭")

# 示例 6：使用 gcf 获取当前图形
def example_gcf():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    fig = plt.gcf()
    fig.suptitle('使用 gcf 获取当前图形')
    plt.show()

# 示例 7：使用 gca 获取当前子图
def example_gca():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    ax = plt.gca()
    ax.set_title('使用 gca 获取当前子图')
    plt.show()

# 示例 8：使用 suptitle 设置图形的总标题
def example_suptitle():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    fig, (ax1, ax2) = plt.subplots(2, 1)
    ax1.plot(x, y1)
    ax1.set_title('子图 1: sin(x)')
    ax2.plot(x, y2)
    ax2.set_title('子图 2: cos(x)')
    fig.suptitle('使用 suptitle 设置图形的总标题')
    plt.show()

# 示例 9：使用 subplots 创建一个包含子图的图形
def example_subplots():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    fig, (ax1, ax2) = plt.subplots(2, 1)
    ax1.plot(x, y1)
    ax1.set_title('子图 1: sin(x)')
    ax2.plot(x, y2)
    ax2.set_title('子图 2: cos(x)')
    plt.show()

# 示例 10：使用 tight_layout 自动调整子图参数以填充图形区域
def example_tight_layout():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    fig, (ax1, ax2) = plt.subplots(2, 1)
    ax1.plot(x, y1)
    ax1.set_title('子图 1: sin(x)')
    ax2.plot(x, y2)
    ax2.set_title('子图 2: cos(x)')
    plt.tight_layout()
    plt.show()

def main():
    print("选择一个示例运行:")
    print("1. 使用 figure 创建一个新的图形")
    print("2. 使用 add_subplot 向图形中添加子图")
    print("3. 使用 savefig 将图形保存到文件")
    print("4. 使用 clf 清除当前图形")
    print("5. 使用 close 关闭图形窗口")
    print("6. 使用 gcf 获取当前图形")
    print("7. 使用 gca 获取当前子图")
    print("8. 使用 suptitle 设置图形的总标题")
    print("9. 使用 subplots 创建一个包含子图的图形")
    print("10. 使用 tight_layout 自动调整子图参数以填充图形区域")
    choice = input("请输入选择的编号 (1-10): ")

    if choice == '1':
        example_figure()
    elif choice == '2':
        example_add_subplot()
    elif choice == '3':
        example_savefig()
    elif choice == '4':
        example_clf()
    elif choice == '5':
        example_close()
    elif choice == '6':
        example_gcf()
    elif choice == '7':
        example_gca()
    elif choice == '8':
        example_suptitle()
    elif choice == '9':
        example_subplots()
    elif choice == '10':
        example_tight_layout()
    else:
        print("无效的选择，请输入 1 到 10 之间的数字。")

if __name__ == "__main__":
    main()
```

## figure函数的参数

| 参数名称       | 功能                           |
| -------------- | ------------------------------ |
| `num`          | 图形编号或名称                 |
| `figsize`      | 图形大小（宽度，高度）         |
| `dpi`          | 每英寸点数（分辨率）           |
| `facecolor`    | 图形背景颜色                   |
| `edgecolor`    | 图形边框颜色                   |
| `frameon`      | 是否绘制图形边框               |
| `tight_layout` | 自动调整子图参数以填充图形区域 |
| `constrained_layout` | 自动调整子图参数以避免重叠 |

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例 1：使用 num 参数设置图形编号或名称
def example_num():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure(num='My Figure')
    plt.plot(x, y)
    plt.title('使用 num 参数设置图形编号或名称')
    plt.show()

# 示例 2：使用 figsize 参数设置图形大小
def example_figsize():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure(figsize=(10, 5))
    plt.plot(x, y)
    plt.title('使用 figsize 参数设置图形大小')
    plt.show()

# 示例 3：使用 dpi 参数设置每英寸点数（分辨率）
def example_dpi():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure(dpi=200)
    plt.plot(x, y)
    plt.title('使用 dpi 参数设置每英寸点数（分辨率）')
    plt.show()

# 示例 4：使用 facecolor 参数设置图形背景颜色
def example_facecolor():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure(facecolor='lightgray')
    plt.plot(x, y)
    plt.title('使用 facecolor 参数设置图形背景颜色')
    plt.show()

# 示例 5：使用 edgecolor 参数设置图形边框颜色
def example_edgecolor():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure(edgecolor='blue')
    plt.plot(x, y)
    plt.title('使用 edgecolor 参数设置图形边框颜色')
    plt.show()

# 示例 6：使用 frameon 参数设置是否绘制图形边框
def example_frameon():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.figure(frameon=False)
    plt.plot(x, y)
    plt.title('使用 frameon 参数设置是否绘制图形边框')
    plt.show()

# 示例 7：使用 tight_layout 参数自动调整子图参数以填充图形区域
def example_tight_layout():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    fig, (ax1, ax2) = plt.subplots(2, 1)
    ax1.plot(x, y1)
    ax1.set_title('子图 1: sin(x)')
    ax2.plot(x, y2)
    ax2.set_title('子图 2: cos(x)')
    plt.tight_layout()
    plt.show()

# 示例 8：使用 constrained_layout 参数自动调整子图参数以避免重叠
def example_constrained_layout():
    x = np.linspace(0, 10, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    fig, (ax1, ax2) = plt.subplots(2, 1, constrained_layout=True)
    ax1.plot(x, y1)
    ax1.set_title('子图 1: sin(x)')
    ax2.plot(x, y2)
    ax2.set_title('子图 2: cos(x)')
    plt.show()

def main():
    print("选择一个示例运行:")
    print("1. 使用 num 参数设置图形编号或名称")
    print("2. 使用 figsize 参数设置图形大小")
    print("3. 使用 dpi 参数设置每英寸点数（分辨率）")
    print("4. 使用 facecolor 参数设置图形背景颜色")
    print("5. 使用 edgecolor 参数设置图形边框颜色")
    print("6. 使用 frameon 参数设置是否绘制图形边框")
    print("7. 使用 tight_layout 参数自动调整子图参数以填充图形区域")
    print("8. 使用 constrained_layout 参数自动调整子图参数以避免重叠")
    choice = input("请输入选择的编号 (1-8): ")

    if choice == '1':
        example_num()
    elif choice == '2':
        example_figsize()
    elif choice == '3':
        example_dpi()
    elif choice == '4':
        example_facecolor()
    elif choice == '5':
        example_edgecolor()
    elif choice == '6':
        example_frameon()
    elif choice == '7':
        example_tight_layout()
    elif choice == '8':
        example_constrained_layout()
    else:
        print("无效的选择，请输入 1 到 8 之间的数字。")

if __name__ == "__main__":
    main()
```
